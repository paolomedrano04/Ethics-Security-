```python
import pandas as pd

# Cambia la ruta si es necesario
df = pd.read_csv('Student_Anonimizacion_Cifrado.csv')

# Muestra las primeras filas para verificar
print(df.head(5))
```

```python
# Encuentra las columnas que terminan en '_cifrado'
cols_cifrado = [col for col in df.columns if col.endswith('_cifrado')]
print("Columnas que debes cifrar luego:", cols_cifrado)
```

```python
# Elimina en el mismo DataFrame
df.drop(columns=cols_cifrado, inplace=True)

# Verifica
print("Columnas actuales:", df.columns.tolist())
print(df.head())
```

```python
# Lista de columnas relevantes para análisis institucional
cols_utiles = [
    'registro_id', 'school', 'age_group', 'sex',
    'Medu', 'Fedu',
    'absences', 'failures_mat', 'failures_por',
    'G1_mat', 'G2_mat', 'G3_mat',
    'G1_por', 'G2_por', 'G3_por',
    'studytime_mat', 'studytime_por',
    'schoolsup_mat', 'schoolsup_por',
    'internet', 'higher', 'paid', 'famsup',
    'traveltime', 'goout', 'Dalc', 'Walc',
    'health', 'freetime'
]

# Filtrar DataFrame
df = df[cols_utiles]

print(df.head())
```

```python
df = df.head(1000)
```

```python
import math

# Total de estudiantes
total_students = len(df)
print(f"Total estudiantes: {total_students}")

# Cada profesor con 100 alumnos
students_per_prof = 100

# Número de profesores necesarios
num_profesores = math.ceil(total_students / students_per_prof)
print(f"Número de profesores: {num_profesores}")

# IDs de profesores
profesor_ids = [f"P{str(i+1).zfill(8)}" for i in range(num_profesores)]

# Asigna ID de profesor y sección
id_profesor_list = []
seccion_list = []

for i in range(total_students):
    prof_idx = i // students_per_prof
    id_profesor_list.append(profesor_ids[prof_idx])
    seccion_list.append(prof_idx + 1)  # Sección = número de profesor

# Agrega columnas
df['id_profesor'] = id_profesor_list
df['seccion'] = seccion_list

# Verifica
print(df[['registro_id', 'id_profesor', 'seccion']].head(120))

# Exporta
df.to_csv('students.csv', index=False, encoding='utf-8')
print("✅ Archivo 'students.csv' exportado con columnas 'id_profesor' y 'seccion'")
```

```python
import pandas as pd
from cryptography.fernet import Fernet

# Generar clave (solo 1 vez)
key = Fernet.generate_key()
cipher = Fernet(key)

print(f'CLAVE: {key.decode()}  --> Guárdala bien!')

# Variables a cifrar
cols_sensitive = ['registro_id', 'school', 'age_group', 'sex', 'id_profesor', 'seccion']

# Función para cifrar strings o ints
def encrypt_value(value):
    value_str = str(value).encode()
    encrypted = cipher.encrypt(value_str)
    return encrypted.decode()

# Aplica a cada columna sensible
for col in cols_sensitive:
    df[f'{col}_enc'] = df[col].apply(encrypt_value)

# Resultado
print(df[[col for col in df.columns if col.endswith('_enc')]].head())

# Guardar si quieres
# df.to_csv('datos_cifrados.csv', index=False)
```

```python
import pandas as pd
import psycopg2

# Configuración de tu base de datos
DB_CONFIG = {
    'host': 'localhost',
    'port': 5432,
    'dbname': 'etica',
    'user': 'postgres',
    'password': '1234'
}

# 1️⃣ Leer el CSV
df = pd.read_csv('students.csv')

# 2️⃣ Conectar a la base de datos
conn = psycopg2.connect(**DB_CONFIG)
cur = conn.cursor()

# 3️⃣ Insertar cada fila
for _, row in df.iterrows():
    cur.execute("""
        INSERT INTO students (
            registro_id, school, age_group, sex, Medu, Fedu,
            absences, failures_mat, failures_por,
            G1_mat, G2_mat, G3_mat, G1_por, G2_por, G3_por,
            studytime_mat, studytime_por, schoolsup_mat, schoolsup_por,
            internet, higher, paid, famsup, traveltime, goout,
            Dalc, Walc, health, freetime, id_profesor, seccion
        )
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s,
                %s, %s, %s, %s, %s, %s, %s, %s, %s, %s,
                %s, %s, %s, %s, %s, %s,
                %s, %s, %s, %s, %s, %s)
    """, tuple(row))

# 4️⃣ Guardar y cerrar
conn.commit()
cur.close()
conn.close()

print("✅ CSV cargado a la tabla 'students' en PostgreSQL.")
```
