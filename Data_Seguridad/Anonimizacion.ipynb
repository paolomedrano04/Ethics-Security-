```python
import pandas as pd

df = pd.read_csv('student_data_cifrado.csv')
df
```

```python
df.columns
```

```python
df['school'] = df['school'].map({'GP': 'A', 'MS': 'B'})
df['sex'] = df['sex'].map({'F': 0, 'M': 1})
df['address'] = df['address'].map({'U': 1, 'R': 0})
df['Pstatus'] = df['Pstatus'].map({'T': 1, 'A': 0})
```

```python
df['age_group'] = pd.cut(df['age'], bins=[14, 17, 20, 23], labels=["15-17", "18-20", "21-22"])
df.drop(columns=["age"], inplace=True)
```

```python
def agrupar_edu(x):
    if x == 0:
        return "Ninguno"
    elif x in [1, 2]:
        return "Básica"
    elif x == 3:
        return "Secundaria"
    else:
        return "Superior"

df['Medu'] = df['Medu'].apply(agrupar_edu)
df['Fedu'] = df['Fedu'].apply(agrupar_edu)

# 4. Agrupar trabajos
def agrupar_trabajo(x):
    if x in ['teacher', 'services']:
        return 'serv_publico'
    elif x == 'health':
        return 'salud'
    else:
        return 'otro'

df['Mjob'] = df['Mjob'].apply(agrupar_trabajo)
df['Fjob'] = df['Fjob'].apply(agrupar_trabajo)
```

```python
columnas = ['registro_id'] + [col for col in df.columns if col != 'registro_id']
df = df[columnas]
```

```python
df.columns
```

```python
anonimizadas = ['school', 'sex', 'age', 'Pstatus', 'address', 'Medu', 'Fedu',
                'Mjob', 'Fjob']

# Construir lista de columnas cifradas que se eliminarán
columnas_a_eliminar = [col + '_cifrado' for col in anonimizadas]

# Eliminar columnas si existen en el DataFrame
df.drop(columns=[col for col in columnas_a_eliminar if col in df.columns], inplace=True)
```

```python
df.columns
```

```python
anonimizadas = ['famsize', 'guardian', 'failures_mat', 'absences', 'G1_mat', 'G2_mat',
                'G3_mat', 'failures_por', 'G1_por', 'G2_por', 'G3_por', 'Dalc', 'Walc',
                'goout', 'romantic', 'internet', 'health']

df.drop(columns=[col for col in columnas_a_eliminar if col in df.columns], inplace=True)
```
