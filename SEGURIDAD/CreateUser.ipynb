```python
import pandas as pd
import random
from faker import Faker

# Inicializar Faker
fake = Faker('es_ES')

# Asegúrate de tener df_users cargado con 'registro_id'
# df_users = pd.read_csv("archivo_con_registro_id.csv")
num_registros = len(df_users)

# Función para generar contraseña temporal
def generar_contraseña():
    return ''.join(random.choices("abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890", k=10))

# Crear listas para los nuevos datos
nombres1, nombres2, apellidos1, apellidos2, correos, contrasenas = [], [], [], [], [], []
correos_existentes = set()

for i in range(num_registros):
    nombre_completo = fake.name().split()

    if len(nombre_completo) >= 4:
        nombre1, nombre2, apellido1, apellido2 = nombre_completo[:4]
    elif len(nombre_completo) == 3:
        nombre1, nombre2, apellido1 = nombre_completo
        apellido2 = fake.last_name()
    else:
        nombre1 = nombre_completo[0]
        nombre2 = fake.first_name()
        apellido1 = fake.last_name()
        apellido2 = fake.last_name()

    base = f"{nombre1.lower()}.{apellido1.lower()}"
    sufijo = ""
    dominio = "@utec.edu.pe"
    correo = base + dominio

    # Paso 1: Agregar letras del segundo apellido si hay duplicados
    for letra in apellido2.lower():
        if correo not in correos_existentes:
            break
        sufijo += letra
        correo = base + sufijo + dominio

    # Paso 2: Si aún es duplicado, agregar números
    contador = 1
    while correo in correos_existentes:
        correo = base + sufijo + str(contador) + dominio
        contador += 1

    # Agregar a set para asegurar unicidad
    correos_existentes.add(correo)

    # Guardar los datos
    nombres1.append(nombre1)
    nombres2.append(nombre2)
    apellidos1.append(apellido1)
    apellidos2.append(apellido2)
    correos.append(correo)
    contrasenas.append(generar_contraseña())

# Asignar columnas al DataFrame
df_users['nombre1'] = nombres1
df_users['nombre2'] = nombres2
df_users['apellido1'] = apellidos1
df_users['apellido2'] = apellidos2
df_users['correo'] = correos
df_users['contraseña'] = contrasenas

print(df_users)
```

```python
# Contar cuántos correos únicos hay
total_correos = df_users['correo'].nunique()
total_registros = len(df_users)

print("Correos únicos:", total_correos)
print("Total de registros:", total_registros)

# Filtrar correos que aparecen más de una vez
correos_duplicados = df_users['correo'].value_counts()
correos_duplicados = correos_duplicados[correos_duplicados > 1]

# Mostrar si existen
if not correos_duplicados.empty:
    print("\n⚠️ Correos repetidos encontrados:")
    print(correos_duplicados)
else:
    print("\n✅ Todos los correos son únicos.")
```

```python
import pandas as pd
import random
from faker import Faker

# Inicializar Faker
fake = Faker('es_ES')

# Reutilizar el df_users ya existente (usuarios estudiantes)
# Supongamos que df_users ya existe con estudiantes

# Set para evitar correos duplicados
correos_existentes = set(df_users['correo'])

# Contadores de IDs por rol
rol_config = {
    "profesor": {"cantidad": 40, "prefijo": "P"},
    "supervisor": {"cantidad": 4, "prefijo": "S"},
    "admin": {"cantidad": 4, "prefijo": "A"}
}

# Función para generar contraseña
def generar_contraseña():
    return ''.join(random.choices("abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890", k=10))

# Función para generar correo con control de duplicados
def generar_correo_unico(nombre, apellido1, apellido2):
    base = f"{nombre[0].lower()}{apellido1.lower()}"
    dominio = "@utec.edu.pe"
    correo = base + dominio

    sufijo = ""
    for letra in apellido2.lower():
        if correo not in correos_existentes:
            break
        sufijo += letra
        correo = base + sufijo + dominio

    contador = 1
    while correo in correos_existentes:
        correo = base + sufijo + str(contador) + dominio
        contador += 1

    correos_existentes.add(correo)
    return correo

# Generar usuarios para cada rol
nuevos_usuarios = []
for rol, info in rol_config.items():
    for i in range(1, info["cantidad"] + 1):
        nombre_completo = fake.name().split()
        
        if len(nombre_completo) >= 4:
            nombre1, nombre2, apellido1, apellido2 = nombre_completo[:4]
        elif len(nombre_completo) == 3:
            nombre1, nombre2, apellido1 = nombre_completo
            apellido2 = fake.last_name()
        else:
            nombre1 = nombre_completo[0]
            nombre2 = fake.first_name()
            apellido1 = fake.last_name()
            apellido2 = fake.last_name()

        correo = generar_correo_unico(nombre1, apellido1, apellido2)
        contraseña = generar_contraseña()
        registro_id = f"{info['prefijo']}{str(i).zfill(8)}"

        nuevos_usuarios.append({
            "registro_id": registro_id,
            "nombre1": nombre1,
            "nombre2": nombre2,
            "apellido1": apellido1,
            "apellido2": apellido2,
            "correo": correo,
            "contraseña": contraseña
        })

# Crear DataFrame con los nuevos usuarios (profesores, supervisores, admins)
df_nuevos = pd.DataFrame(nuevos_usuarios)
```

```python
import unicodedata

# Función para remover tildes y acentos
def quitar_tildes(texto):
    if isinstance(texto, str):
        return ''.join(
            c for c in unicodedata.normalize('NFKD', texto)
            if not unicodedata.combining(c)
        )
    return texto

# Aplicar a columnas relevantes en df_users y df_nuevos
columnas_a_limpieza = ['nombre1', 'nombre2', 'apellido1', 'apellido2', 'correo']

for col in columnas_a_limpieza:
    df_users[col] = df_users[col].apply(quitar_tildes)
    df_nuevos[col] = df_nuevos[col].apply(quitar_tildes)
```

```python
# Unir ambos DataFrames
df_final = pd.concat([df_users, df_nuevos], ignore_index=True)

# Guardar en CSV
df_final.to_csv("usuarios_completos.csv", index=False)
print("✅ Archivo guardado como usuarios_completos.csv sin tildes.")
```
